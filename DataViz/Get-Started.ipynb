## Ticks

````python
fig, ax = plt.subplots(1, 2, figsize=(16, 4))

ax[0].set_xlim(0, 10)
ax[0].set_ylim(0, 100)

ax[1].set_xlim(0, 10)
ax[1].set_ylim(0, 100)

ax[1].set_xticks([0, 2, 4, 6, 8, 10])
ax[1].set_yticks([0, 20, 40, 60, 80, 100])

# Use set_xticklabels -> após setar os ticks com set_xticks 
ax[1].set_xticklabels(['Zero', 'Two', 'Four', 'Six', 'Eight', 'Ten'])
ax[1].set_yticklabels(['0%', '20%', '40%', '60%', '80%', '100%'])

plt.show()
````

#### Mas qual é a diferença entre o *ticks* e o *ticklabels*?

✅ **ticks (marcadores)**  
São as posições numéricas ao longo dos eixos onde você quer que apareçam os valores ou rótulos.
- Pense neles como os "pauzinhos" (ou "sticks") que marcam um ponto no eixo.
- Eles indicam onde alguma informação será mostrada no eixo (mesmo que sem texto)
```python
ax.set_xticks([0, 2, 4, 6, 8, 10])  # Isso define que haverá marcas nesses pontos ao longo do eixo X.
```

✅ **ticklabels (rótulos dos marcadores)**  
São os textos que aparecem ao lado dos ticks. Eles representam o que será mostrado como rótulo em cada uma das posições dos ticks.
```python
ax.set_xticklabels(['Zero', 'Dois', 'Quatro', 'Seis', 'Oito', 'Dez'])  # Isso define que os ticks definidos anteriormente terão esses rótulos personalizados.
```

## Legends

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [2, 4, 1], label='Linha A')
ax.plot([1, 2, 3], [1, 2, 3], label='Linha B')
ax.legend()  # Exibe a legenda automaticamente
plt.show()
````

#### Posições da **legenda**

📌 Principais opções:
- `'best'` (padrão): escolhe a melhor posição automática
- `'upper right'`, `'upper left'`
- `'lower right'`, `'lower left'`
- `'center right'`, `'center left'`
- `'upper center'`, `'lower center'`
- `'center'`

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [2, 4, 1], label='Linha A')
ax.plot([1, 2, 3], [1, 2, 3], label='Linha B')
ax.legend(loc='upper left')
plt.show()
````

#### Legenda fora do gráfico
- `bbox_to_anchor` & `loc`

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [2, 4, 1], label='Linha A')
ax.plot([1, 2, 3], [1, 2, 3], label='Linha B')

ax.legend(loc='upper right', bbox_to_anchor=(1.25, 1))  # canto superior direito
ax.legend(loc='upper left', bbox_to_anchor=(-0.35, 1))  # canto superior esquerdo
ax.legend(loc='lower right', bbox_to_anchor=(1.25, 0))  # canto inferior direito
ax.legend(loc='lower left', bbox_to_anchor=(-0.35, 0))  # canto inferior esquerdo

plt.show()
```

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3], [2, 4, 1], label='Linha A')
ax.plot([1, 2, 3], [1, 2, 3], label='Linha B')

ax.legend(loc='lower right', bbox_to_anchor=(1.25, 0))

plt.show()
````

Mas como isso funciona, realmente funciona:

1. Ajuste os valores de `bbox_to_anchor`:
    * Valores >1 movem para fora à direita/acima.
    * Valores <0 movem para fora à esquerda/abaixo.
2. Combinação de loc e bbox_to_anchor:  
    - `loc` define qual ponto da legenda será posicionado em `bbox_to_anchor`.
3. Se a legenda ainda estiver cortada:
```python
   - plt.subplots_adjust(right=0.75)  # Libera espaço à direita
```

#### Personalizando sua legenda
- `Patch`: Formas geometricas ~> representa blocos preenchidos (ex: barras, áreas).
- `Line2D`: Linhas ~> representa linhas ou pontos (ex: gráficos de linha, scatter).

O parâmetro `label`='...' em cada objeto define o texto da legenda.  
O `handles`=... na `ax.legend()` aceita qualquer lista mista de objetos gráficos.

```python
# Dessa forma você cria a sua legenda da maneira que achar melhor
custom_legend_items = [
    Patch(facecolor='green', label='Classe A'),
    Patch(facecolor='orange', label='Classe B'),
    Line2D([0], [0], color='red', lw=2, linestyle='--', label='Tendência'),
    Line2D([0], [0], color='blue', lw=2, marker='o', linestyle='', label='Amostra')
]

# Utilize o handles
ax.legend(handles=custom_legend_items)
```

````python
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Sua legenda customizada
custom_legend_items = [
    Patch(facecolor='green', label='Classe A'),
    Patch(facecolor='orange', label='Classe B'),
    Line2D([0], [0], color='red', lw=2, linestyle='--', label='Tendência'),
    Line2D([0], [0], color='blue', lw=2, marker='o', linestyle='', label='Amostra')
]

# Passar todos como "handles" para a legenda
ax.legend(handles=custom_legend_items)

plt.show()
````

## Line2D, Patch, Text

`Line2D`: linhas customizadas;  
`Patch`: formas como retângulos, círculos, polígonos, etc...  
`Collection`: grupos eficientes de elementos gráficos.

#### Line2D

````python
from matplotlib.lines import Line2D
# Use para anotações técnicas ou layouts sob medida e linhas completamente customizadas
fig, ax = plt.subplots()
line = Line2D([0, 1], [0, 1], c='red', linewidth=2, linestyle='--')
ax.add_line(line)
ax.set_xlim(-1, 2)
ax.set_ylim(-1, 2)
plt.show()
````

#### Patch

````python
from matplotlib.patches import Rectangle, Circle
# Use para destacar áreas
fig, ax = plt.subplots()
rect = Rectangle((1, 1), width=2, height=1, color='skyblue', alpha=0.7)
ax.add_patch(rect)

circle = Circle((3, 2), radius=0.5, color='orange', alpha=0.5)
ax.add_patch(circle)

ax.set_xlim(0, 5)
ax.set_ylim(0, 4)
plt.gca().set_aspect('equal')
plt.show()
````

#### Collection

````python
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
# Use para customização e automação
fig, ax = plt.subplots()

patches = []
for i in range(10):
    circle = Circle((i, i), radius=0.3)
    patches.append(circle)

collection = PatchCollection(patches, color='green', alpha=0.5)
ax.add_collection(collection)

ax.set_xlim(-1, 10)
ax.set_ylim(-1, 10)
plt.gca().set_aspect('equal')
plt.show()
````

## Eixos Duplos (twinx() & twiny())

````python
# Dados
x = np.linspace(0, 10, 100)
y1 = np.sin(x) * 20 + 25  # Temperatura (°C)
y2 = np.cos(x) * 30 + 50  # Umidade (%)
````

````python
# Figura e eixo principal
fig, ax = plt.subplots()
ax.plot(x, y1, 'r-', label='Temperatura (°C)')
ax.set_xlabel('Tempo (h)')
ax.set_ylabel('Temperatura (°C)', color='r')
ax.tick_params(axis='y', labelcolor='r')

# Eixo Y secundário (twinx)
ax2 = ax.twinx()
ax2.plot(x, y2, 'b--', label='Umidade (%)')
ax2.set_ylabel('Umidade (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Legenda unificada
lines1, labels1 = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax.legend(lines1 + lines2, labels1 + labels2)

plt.title('Temperatura vs Umidade')
plt.show()
````

````python
# Dados
x_celsius = np.linspace(0, 100, 10)
y = x_celsius ** 2  # Dummy data

# Figura e eixo principal (Celsius)
fig, ax = plt.subplots()
ax.plot(x_celsius, y, 'g-', label='Celsius')
ax.set_xlabel('Temperatura (°C)')
ax.set_ylabel('Energia (J)')

# Eixo X secundário (twiny) - Fahrenheit
ax2 = ax.twiny()
x_fahrenheit = x_celsius * 9/5 + 32  # Conversão
ax2.plot(x_fahrenheit, y, 'm--', label='Fahrenheit')  # Mesmo Y, X diferente
ax2.set_xlabel('Temperatura (°F)')

# Linhas verticais para destacar pontos
ax.axvline(0, color='gray', linestyle=':')
ax2.axvline(32, color='gray', linestyle=':')  # 0°C = 32°F

plt.title('Relação Energia vs Temperatura (Celsius/Fahrenheit)')
plt.legend()
plt.show()
````

````python
fig, ax = plt.subplots()

# Eixo principal (inferior/esquerda)
ax.plot(x, y1, 'r-', label='Dados 1 (Eixo Principal)')
ax.set_xlabel('X Principal')
ax.set_ylabel('Y Principal', color='r')
ax.tick_params(axis='y', labelcolor='r')

# Eixo Y secundário (direita)
ax2 = ax.twinx()
ax2.plot(x, y2, 'b--', label='Dados 2 (Eixo Y Secundário)')
ax2.set_ylabel('Y Secundário', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Eixo X secundário (topo)
ax3 = ax.twiny()
ax3.plot(x * 2, y1, 'g:', label='Dados 3 (Eixo X Secundário)')  # Escala X diferente
ax3.set_xlabel('X Secundário')

plt.title('Gráfico com Eixos Duplos em X e Y')
fig.legend(loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=3)
plt.tight_layout()
plt.show()
````
